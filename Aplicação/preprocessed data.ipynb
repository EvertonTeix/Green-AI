como essas 4 features contém muitos valores nulos, vamos pagar elas.

```python
base_dados = base_dados.drop(columns=['ContractEndDate', 'WorkExperience', 'ActiveLateCategory', 'WorseLateCategory'])
```

```python
base_dados
```

apagando as features (agora só restam as numércias) que possuem mais de 40% dos valores ausentes.

```python
nullFeatures = []
for col in base_dados.columns:
    nullValues = base_dados[col].isnull().sum()
    nullPercentage = nullValues/base_dados.shape[0]
    if nullPercentage > 0.4:
        print(f"{col} : {nullPercentage*100:.2f}%")
        nullFeatures.append(col)
print(nullFeatures)
```

```python
remover_quarenta = ['PlannedPrincipalTillDate', 'PlannedPrincipalPostDefault', 'PlannedInterestPostDefault', 'EAD1', 'EAD2', 'PrincipalRecovery', 'InterestRecovery', 'PrincipalWriteOffs', 'InterestAndPenaltyWriteOffs', 'PreviousEarlyRepaymentsBefoleLoan', 'NextPaymentNr', 'NrOfScheduledPayments', 'PrincipalDebtServicingCost', 'InterestAndPenaltyDebtServicingCost']

base_dados = base_dados.drop(columns=remover_quarenta)
```

vendo a quantidade de nulos das features numericas

```python
null_count_num = base_dados.isnull().sum()
print(null_count_num[null_count_num > 0])
```

```python
base_dados.info()
```

separando em numericos e categoricos, como categoricos so existe a 'Status', que é a classe, entao ficou separado em classe e numericos

```python
classe = base_dados['Status']
numericos = base_dados.drop(['Status'], axis=1)
```

aqui estou preenchendo os valores nulos pela media da sua respectiva coluna

```python
numericos = numericos.fillna(numericos.median())
null_count_num = numericos.isnull().sum()
print(null_count_num[null_count_num > 0])
```

```python
classe
```

```python
numericos
```

aqui estou juntando a classe com os numericos pra assim ter a base de dados de volta

```python
base_dados = numericos.join(classe)
```

```python
base_dados
```

a feature NewCreditCustomer era do tipo bool, então mudei para numerico

```python
base_dados['NewCreditCustomer'] = base_dados['NewCreditCustomer'].astype(int)
base_dados["NewCreditCustomer"].value_counts()
```

```python
base_dados.info()
```

como vamos prever o risco de inadimplência ou não inadimplência, não fazia sentido ter as amostras 'Current', uma vez que elas não dão uma certeza de pago ou não pago. Então removemos todas as amostras deste tipo.

```python
currentLoans = base_dados[base_dados["Status"]=="Current"]
base_dados = base_dados.drop(currentLoans.index)
base_dados.shape
```

```python
base_dados["Status"].value_counts()
```

Como o computador não entende palavras, renomeamos a classe para 0 (Repaid) e 1 (Late)

```python
base_dados['Status'] = base_dados['Status'].map({'Repaid': 0, 'Late': 1}) # 0 -> pago; 1-> não pago
```

```python
base_dados["Status"].value_counts()
```
